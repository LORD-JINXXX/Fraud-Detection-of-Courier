#Function For Assigning Weight Slab

```python
def weight_slab(weight):
    i=round(weight%1,1)
    if i==0.0:
        return weight
    elif i>0.5:
        return int(weight)+1.0
    else:
        return int(weight)+0.5
```

#Generation Of Weight Slab Column

```python
##Generation of Weight Slab
re_df_sku_inorder_report['Weight Slab As Per X (KG)']=re_df_sku_inorder_report['Total Weight AS Per X (KG)'].apply(weight_slab)
df_courier_invoice['Weight Slab Charged By Courier Company (KG)']=(df_courier_invoice['Total Weight As Per Courier Company (KG)']).apply(weight_slab)
```

#Computation Of Expected Charge

```python
total_Expected_charge=[]
for i in range(len(re_df_sku_inorder_report)):
    fwd_category='fwd_'+re_df_sku_inorder_report['Delivery Zone As Per X'][i]
    fwd_fixed=df_courier_rates[fwd_category+'_fixed'][0]
    fwd_additional=df_courier_rates[fwd_category+'_additional'][0]
    rto_category='rto_'+re_df_sku_inorder_report['Delivery Zone As Per X'][i]
    rto_fixed=df_courier_rates[rto_category+'_fixed'][0]
    rto_additional=df_courier_rates[rto_category+'_additional'][0]
    if re_df_sku_inorder_report['Type of Shipment'][i]=='Forward charges':
        if re_df_sku_inorder_report['Weight Slab As Per X (KG)'][i]<=0.5:
            total_Expected_charge.append(fwd_fixed)
        else:
            additional_weight=(re_df_sku_inorder_report['Weight Slab As Per X (KG)'][i]-0.5)/0.5
            total_Expected_charge.append(fwd_fixed+additional_weight*fwd_additional)
    if re_df_sku_inorder_report['Type of Shipment'][i]=='Forward and RTO charges':
        if re_df_sku_inorder_report['Weight Slab As Per X (KG)'][i]<=0.5:
            total_Expected_charge.append(fwd_fixed+rto_fixed)
        else:
            additional_weight=(re_df_sku_inorder_report['Weight Slab As Per X (KG)'][i]-0.5)/0.5
            total_Expected_charge.append(fwd_fixed+additional_weight*(fwd_additional+rto_additional))

re_df_sku_inorder_report['Expected Charge As Per X (Rs.)']=total_Expected_charge
```

#Computation Of Difference Between Expected Charges And Billed Charges

```python
diff_bw_ec_bc="""
Difference_Between_Expected_Charges_and_Billed_Charges_Rs =(Expected_Charge_As_Per_X_Rs - Charges_Billed_By_Courier_Company_Rs)
"""
resultant_output1.eval(diff_bw_ec_bc,inplace=True,engine='python')
```

```python
resultant_output1=resultant_output1.rename(columns={'Charges_Billed_By_Courier_Company_Rs':'Charges Billed By Courier Company (Rs.)','Expected_Charge_As_Per_X_Rs':'Expected Charge As Per X (Rs.)','Difference_Between_Expected_Charges_and_Billed_Charges_Rs':'Difference Between Expected Charges And Billed Charges (Rs.)'})
```

```python
writer = pd.ExcelWriter('Expected_Result.xlsx',if_sheet_exists="replace",engine='openpyxl',mode="a")
```

```python
resultant_output1.to_excel(writer,sheet_name='Calculations')
calculations=writer.save()
```

#Computation Of Total Orders

```python
correctly_charged=resultant_output1[resultant_output1['Difference Between Expected Charges And Billed Charges (Rs.)']==0.0]
correctly=["Total Orders where X has been correctly charged",len(correctly_charged),sum(correctly_charged['Charges Billed By Courier Company (Rs.)'])]

over_charged=resultant_output1[resultant_output1['Difference Between Expected Charges And Billed Charges (Rs.)']<0.0]
over=["Total Orders where X has been over charged",len(over_charged),np.absolute(sum(over_charged['Difference Between Expected Charges And Billed Charges (Rs.)']))]

under_charged=resultant_output1[resultant_output1['Difference Between Expected Charges And Billed Charges (Rs.)']>0.0]
under=["Total Orders where X has been under charged",len(under_charged),np.absolute(sum(under_charged['Difference Between Expected Charges And Billed Charges (Rs.)']))]
```

```python
resultant_output2=pd.DataFrame([correctly,over,under],columns=['Description','Count','Amount (Rs.)'])
```
